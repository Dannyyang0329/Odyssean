```python
import json
iframe_data = json.load(open('../data/iframe_urls.json'))
intro_data = json.load(open('../data/attractions_intro.json'))
attraction_list1 = list(intro_data.keys())
attraction_list2 = list(iframe_data.keys())
```

```python
and_list = list(set(attraction_list1) & set(attraction_list2))
and_list, len(and_list)
```

```python
new_data = {}
ori_attraction_data = json.load(open('../data/attractions.json'))
for country in ori_attraction_data:
    new_data[country] = []
    for attraction in ori_attraction_data[country]:
        attraction_name = attraction['name']
        if attraction_name in and_list:
            new_data[country].append({
                "name": attraction['name'],
                "image_url": attraction['image_url'],
                "score": attraction['score'],
                "iframe_url": iframe_data[attraction_name],
                "intro": intro_data[attraction_name]
            })
```

```python
with open('../data/attractions_with_iframe.json', 'w') as f:
    json.dump(new_data, f, indent=4)
```
